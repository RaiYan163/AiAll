```python
def remove_punctuation(sentence):
    for x in punct:
        sentence = sentence.replace(x, "")
    return sentence

for example in dataset1:
    example["content"] = remove_punctuation(example["content"])
```

```python
def punct_sw_rem(sentence):

    for p in punct:
        sentence["content"] = sentence["content"].replace(p, " ")

    words = sentence["content"].split()

    sentence["content"] = ""
    for word in words:
        if word not in sample_stop_words:
            sentence["content"] += word + " "

    sentence["content"] = sentence["content"].strip()

    return sentence
```

```python
sample = dataset1.select([i for i in range(25)])
sample
```

```python
preprocessed_dataset = sample.map(punct_sw_rem)
sample[:23]
```

```python
filtered_dataset = preprocessed_dataset.filter(lambda sentence: len(sentence["content"].split()) >= 10)
filtered_dataset[:10]
```

```python
shards = []
for i in range(4):
  shards.append(filtered_dataset.shard(num_shards = 5, index = i))
print(shards)
```
